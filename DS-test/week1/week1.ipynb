### Counting Columns 

Here I calculated the number of occurrences of each of the different values of ‘Gender’ and ‘Caste’ in order to have a more intuitive understanding of the distribution of male and female students and the distribution of the number of students with different castes.

### Sorting

Here, the top 10 best performing students are sorted according to the ‘Performance’ column, making it easy to quickly identify high scorers.

```python
df.sort_values(by='Performance', ascending=False)[:10]
```

### Filtering

Here I've selected all the rows in the coaching column that have a value other than ‘NO’, that is, they will contain ‘yes’, null, or any other value that is not ‘NO’.

```python
#Filtering of students for coaching
coached_students = df[df["coaching"] != "NO"]
coached_students
```

### Missing Values

Data integrity and missing data is checked here. Finding empty records in the ‘time’ column helps to identify possible missing issues in the data.

```python
#Divide without NaN by total length revealing which columns have missing values
df.count() / len(df)
```

```python
#Look at rows for which BuildingName(s) contains NaN
df[df["time"].isna()]["time"].values
```
